```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('data.csv')
data = data.set_index('id')
data = data.iloc[:,:-1]

features = data.columns[2:12]

bins = 12 #no of subpoints in the graph

plt.figure(figsize = (15,15)) #initialize the graph, see what num you like for space 15x15 is the total plot, in which 12 graph is plotted 
#start counter, everytime graph is created-> id increases

for i,feature in enumerate(features):
    rows = int((len(features)+1/2)) #typecasting here:int, +1 cause if odd no of graph, align them all, 2 is the no of columns here: fixed
    plt.subplot(rows,2,i+1) #2-columns,i+1 graph starts from 1, i starts from 0
    sns.distplot(data[data.diagnosis == 'M'][feature],bins=bins,color='red',label='Malignant') 
    sns.distplot(data[data.diagnosis == 'B'][feature],bins=bins,color='blue',label='Benign')
    plt.legend(loc='upper right')
    #plt.tight_layout() #tightlayout is used to remove the white spaces
    plt.show()
```
